```python
# To display graphs in this notebook, run this:
%matplotlib inline

# We're now going to import some important libraries

import numpy.random # for generating a noisy data set
from sklearn import linear_model # for training a linear model

import matplotlib.pyplot as plt # for plotting in general
from math import log
import math

import pandas as pd
```

```python
#Load Data
data = pd.read_csv("kmer_graphs_data.csv")
data.head()
```

```python
#Define Values
k = list(data['k'])

listofall = []

listofall.append(list(data['Estimated k-mers'])) #eksmers
listofall.append(list(data['4^k'])) #bound4k
listofall.append(list(data['Number of Distinct k-mers']))#len(genome)-k+1
```

```python
#Make Log values

def logoflist(listz):
    loglist = []
    for i in listz:
        loglist.append(math.log(i, 10))
    return loglist

loglistofall = []

for i in listofall:
    loglistofall.append(logoflist(i))
```

```python
#Make graphs

fig = matplotlib.pyplot.figure(1)
fig.suptitle('Estimated Distinct k-mers from Jellyfish Count')
plt.xlabel('K')
plt.ylabel('Distinct K-mers')
#plt.plot(k, loglistofall[0], loglistofall[1])

plt.plot(k, loglistofall[0], "-b", label="Jellyfish Count Distinct K-mers")
plt.plot(k, loglistofall[1], "-r", label="4^K Distinct K-mers")
plt.plot(k, loglistofall[2], "-g", label="len(Genome)-K+1 Distinct K-mers")
plt.legend(loc="upper left")
plt.show()
```

```python
#Higher Resolution?
fig = matplotlib.pyplot.figure(1)
fig.suptitle('Estimated Distinct k-mers')
matplotlib.pyplot.xlabel('K')
matplotlib.pyplot.ylabel('Distinct K-mers')

plt.plot(k, loglistofall[2], "-g", label="len(Genome)-K+1 Distinct K-mers")
plt.plot(k, loglistofall[1], "-y", label="4^K Distinct K-mers")
plt.legend(loc="upper left")
plt.show()
```
