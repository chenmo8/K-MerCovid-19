```python
# We're going to be doing some plotting, and we want to be able to see these plots.
# To display graphs in this notebook, run this cell.
%matplotlib inline

# We're now going to import some important libraries

import numpy.random # for generating a noisy data set
from sklearn import linear_model # for training a linear model

import matplotlib.pyplot # for plotting in general
from math import log
import math

import pandas as pd
```

```python
'''
File is 'EPI_ISL_402128.fasta'
'''

#genome = open("EPI_ISL_402128.fasta", "r")
#readgenome = genome.read()

readgenomelen = 30278

'''
for letter in readgenome:
    print(letter)
'''

listofpossiblekmers = {} #key is k, value is #of distinct k-mers
for k in range(0, 31):
    distinctkmer = readgenomelen-k+1
    listofpossiblekmers.update({k:distinctkmer})

print(listofpossiblekmers.values())   

loggedk = []
kmervalues = []
for i in list(listofpossiblekmers.keys()):
    if i == 0:
        loggedk.append(0)
    else:
        loggedk.append(math.log(i, 10))
for i in list(listofpossiblekmers.values()):
    kmervalues.append(math.log(i, 10))
    
```

```python
fig = matplotlib.pyplot.figure(1)
fig.suptitle('Distinct K-mers')
matplotlib.pyplot.xlabel('K')
matplotlib.pyplot.ylabel('Distinct K-mers')
matplotlib.pyplot.plot(list(listofpossiblekmers.keys()), kmervalues)#len(genome)-k+1
#matplotlib.pyplot.plot(loggedk, fourlogged)#4^k
```

```python
fourdistinctkmers = {}
for k in range(0, 31):
    distinctkmer = 4^k
    fourdistinctkmers.update({k:distinctkmer})
print(fourdistinctkmers.values())
```

```python
#List of Distinct K-mers, if it is equal to 4^K
fourlogged = []
for i in list(fourdistinctkmers.values()):
    if i == 0:
        fourlogged.append(i)
    else:
        fourlogged.append(math.log(i, 10))
```

```python
fig = matplotlib.pyplot.figure(1)
fig.suptitle('Distinct K-mers = 4^k on Log Scale')
matplotlib.pyplot.xlabel('K')
matplotlib.pyplot.ylabel('Distinct K-mers')
matplotlib.pyplot.plot(list(listofpossiblekmers.keys()), fourlogged)
```
